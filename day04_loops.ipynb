```python
import random
import time

love_answers = ["Romance is in your future", "Guard your heart", "Love will find you"]
money_answers = ["Wealth awaits", "Save your coins", "A windfall approaches"]
general_answers = ["Yes", "No", "Maybe", "The future is unclear"]

print("ðŸ”® Welcome to the Fortune Teller!")
print("Type 'quit' to leave.\n")

while True:
    question = input("Ask your question: ").lower()

    if question == "quit":
        print("The Fortune Teller fades into the mist...")
        break

    print("The mists swirl...")
    time.sleep(2)

    if "love" in question:
        print(f"The Fortune Teller says: {random.choice(love_answers)}")
    elif "money" in question:
        print(f"The Fortune Teller says: {random.choice(money_answers)}")
    else:
        print(f"The Fortune Teller says: {random.choice(general_answers)}")

    print()  # blank line between questions
```
