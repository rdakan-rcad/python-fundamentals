```python
# Quiz Engine v1 - Basic Setup

questions = [
    "What keyword exits a while loop?",
    "What index is the first item in a list?",
    "What method adds an item to the end of a list?"
]

answers = ["break", "0", "append"]

score = 0

user_input = input(questions[0] + " ").lower()

if user_input == answers[0]:
    score = score + 1
    print("Correct!")
else:
    print(f"Sorry, the answer was: {answers[0]}")

print(f"Score: {score}")
```

```python
questions = [
    "What keyword exits a while loop?",
    "What index is the first item in a list?",
    "What method adds an item to the end of a list?"
]

answers = ["break", "0", "append"]

score = 0
index = 0

for q in questions:
    user_input = input(q + " ").lower()

    if user_input == answers[index]:
        score = score + 1
        print("Correct!")
    else:
        print(f"Sorry, the answer was: {answers[index]}")

    index = index + 1

print(f"Quiz complete! You got {score} out of {len(questions)} correct.")
```
