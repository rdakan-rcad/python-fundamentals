```python
def greet(name, greeting="Hello"):
    print(f"{greeting}, {name}!")

greet("Rick")
greet("Rick", "Good morning")
```

```python
def announce_loot(item, quantity=1):
  print(f"You found {quantity} {item}!")

announce_loot("sword")
announce_loot("arrows", 10)
```

```python
def add_bonus(score):
    score = score + 10
    print(f"Inside function: {score}")

player_score = 50
add_bonus(player_score)
print(f"Outside function: {player_score}")
```

```python
def add_bonus(score):
    return score + 10

player_score = 50
player_score = add_bonus(player_score)
print(player_score)  # now it's 60
```

```python
def announce_item(item, quantity):
    print(f"You picked up {quantity} {item}!")

announce_item("arrows", 5)
announce_item("gold", 100)
```

```python
def add_numbers(a, b):
    return a + b

result = add_numbers(5, 3)
print(result)

total = add_numbers(10, 20) + add_numbers(1, 2)
print(total)
```

```python
def show_inventory(inventory):
    print(f"=== YOUR INVENTORY ===")
    for item in inventory:
        print(f"{item}: {inventory[item]}")

inventory = {"arrows": 10, "gold": 50}
show_inventory(inventory)
```
