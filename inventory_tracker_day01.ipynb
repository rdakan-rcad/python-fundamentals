```python
inventory = {"sword": 0, "bow": 0, "arrows": 0, "gold": 100}

while True:
    print("\nWhat would you like to do?")
    print("1. View inventory")
    print("2. Add item")
    print("3. Quit")

    choice = input("Enter choice: ")

    if choice == "1":
        for item in inventory:
          print(f"You have: {item} - {inventory[item]}")
    elif choice == "2":
      item_name = input("What item? ")
      item_qty = int(input("How many? "))
      inventory[item_name] = inventory[item_name] + item_qty
    elif choice == "3":
        print("Goodbye!")
        break



```
