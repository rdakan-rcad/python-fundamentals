```python
import random

def attempt_action(days_spent):
    """Try an action. More days = better chance of success."""
    roll = random.randint(1, 5)
    success = roll <= days_spent
    return success

def adjust_confidence(direction):
    """Move confidence up or down. Direction is 'up' or 'down'."""
    current = confidence_levels.index(player["confidence"])

    if direction == "up" and current < 4:
        player["confidence"] = confidence_levels[current + 1]
    elif direction == "down" and current > 0:
        player["confidence"] = confidence_levels[current - 1]

confidence_levels = ["very low", "low", "medium", "high", "very high"]

player = {
    "time": 30,
    "skills": [],
    "confidence": confidence_levels[2]
}

print("=" * 50)
print("OTTO BLISS: THE BEGINNING")
print("=" * 50)
print()
print("You're sitting in the campus coffee shop, sketchbook open.")
print("Your latest Otto Bliss illustration stares back at you.")
print()
print("Your friend slides into the seat across from you.")
print('"So," they say, "what are you actually going to DO with Otto Bliss?"')
print()
print("You've been thinking about this. You have 30 days until")
print("the end of semester showcase. Time to pick a direction.")
print()
print("What do you want to explore?")
print("1. Build a web presence")
print("2. Create a game featuring Otto")
print("3. Launch a merch line")

choice = input("Enter 1, 2, or 3: ")

days_spent = int(input("How many days will you commit? (1-5): "))
player["time"] = player["time"] - days_spent

success = attempt_action(days_spent)

if choice == "1":
    print()
    print("Web presence... yes. Otto Bliss needs to exist online.")
    if success:
        print("You make real progress! Your portfolio site is taking shape.")
        player["skills"].append("web research")
        adjust_confidence("up")
    else:
        print("You go down a rabbit hole of tutorials and end up confused.")
        adjust_confidence("down")

elif choice == "2":
    print()
    print("A game! Interactive storytelling could bring Otto to life.")
    if success:
        print("You find a beginner-friendly engine and build a simple prototype!")
        player["skills"].append("game design basics")
        adjust_confidence("up")
    else:
        print("The game engine documentation overwhelms you. Nothing to show yet.")
        adjust_confidence("down")

elif choice == "3":
    print()
    print("Merch. People wearing Otto Bliss out in the world.")
    if success:
        print("You set up a print-on-demand account and upload your first design!")
        player["skills"].append("merch research")
        adjust_confidence("up")
    else:
        print("Pricing and shipping options make your head spin. No progress today.")
        adjust_confidence("down")

else:
    print("You stare blankly at your friend. They wave a hand in front of your face.")

print()
print("=" * 50)
print("END OF SCENE 1")
print("=" * 50)
print()
print(f"Time remaining: {player['time']} days")
print(f"Skills learned: {player['skills']}")
print(f"Confidence: {player['confidence']}")
```
